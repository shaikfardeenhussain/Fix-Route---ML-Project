```python

def NearestTechnicianEstimate():

    print("=== Nearest Technician Estimator for Your Dataset ===")

    # Load your Enhanced dataset
    inpdata = pd.read_csv("decfin.csv")
    print("Sample of Input Data:")
    print(inpdata.head())
    print("\n==============================================================\n")

    print("Available User Locations:")
    unique_locs = inpdata["User_Location"].dropna().unique()
    print(unique_locs)

    user_loc = input("\nEnter User Location exactly as shown: ").strip()

    print("\nProcessing...\n")

    # Create temp copy
    temp = inpdata.copy()

    # Replace all rows' user location with user input
    temp["User_Location"] = user_loc

    # Load encoders and model
    ohe = joblib.load("finalohe.pkl")
    sc = joblib.load("finalsc.pkl")
    model = joblib.load("finalsvr.pkl")   # or "xgb.pkl"

    # Your categorical columns
    ohe_cols = ["User_Location", "Technician_Name", "Technician_Location"]

    # Apply One-Hot Encoding
    temp_ohe = ohe.transform(temp[ohe_cols]).toarray()
    temp_ohe = pd.DataFrame(temp_ohe, columns=ohe.get_feature_names_out())

    # Numeric columns
    num_cols = ["Technician_Rating", "Base_Charge", 
                "User_Lat", "User_Lng", "Tech_Lat", "Tech_Lng"]

    temp_num = temp[num_cols]

    # Apply scaling
    temp_num_scaled = pd.DataFrame(
        sc.transform(temp_num),
        columns=num_cols
    )

    # Combine numeric + encoded categorical
    X_real = pd.concat([temp_num_scaled, temp_ohe], axis=1)

    # Align columns to match model training
    model_cols = model.feature_names_in_

    # Add missing Feature columns
    for col in model_cols:
        if col not in X_real.columns:
            X_real[col] = 0

    # Sort X_real by model's expected column order
    X_real = X_real[model_cols]

    # Predict distance
    temp["Predicted_Distance_KM"] = model.predict(X_real)

    # Sort technicians by nearest predicted distance
    nearest = temp.sort_values("Predicted_Distance_KM").head(3)

    print("=== TOP 3 NEAREST TECHNICIANS ===\n")
    for _, row in nearest.iterrows():
        print("Technician:", row["Technician_Name"])
        print("Location:", row["Technician_Location"])
        print("Rating:", row["Technician_Rating"])
        print("Base Charge:", row["Base_Charge"])
        print("Predicted Distance:", round(row["Predicted_Distance_KM"], 2), "km")
        print("--------------------------------------------------")

    return nearest
```
